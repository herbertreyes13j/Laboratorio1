## Modelo de GIT

Git modela el historia de una coleccion de archivos y carpetas dentro de una carpeta de nivel superior como una serie de instanantenas. Para git un archivo se llama "blob" y es simplemente un monton de bytes, mientras que un directorio se le llama "arbol" y mapea nombres a otros blobs o arboles. Y finalmente una snapshot es un arbol de nivel superrior que se esta monitoreando. Como por ejemplo:

```
<root> (tree)
|
+- foo (tree)
|  |
|  + bar.txt (blob, contents = "hello world")
|
+- baz.txt (blob, contents = "git is wonderful")

```


El arbol superior contiene dos elementos, un arbol llamado "foo", que este contiene un archivo llamando bar.txt y un archivo llamado baz.txt. 


En git un historial es un grafo aciclico dirigido (DAG) de instananeas, lo que significa que GIT hace una referencia a un conjunto de "padres" de snapshots que le precedieron, ya que un snapshot podria descender de multiples padres si se diera el merge al combinar dos ramas, como se muestra en el siguiente ejemplo:

```
o <-- o <-- o <-- o
            ^
             \
              --- o <-- o

```

A estos snapshots git les da el nombre de commits. 

## Referencias

Todos los snapshots se identifican por su codigo hash SHA-1, lo que para los humanos no puede ser muy facil de entender al ser cadenas muy largas, por lo que la forma en que GIT ha solucionado este problema es por medio de los nombres legibles para humanos para los hashes, llamadas "Referencias", que son punteros a estos. Un ejemplo en codigo de como puede ser visto es el siguiente:

```
references = map<string, string>

def update_reference(name, id):
    references[name] = id

def read_reference(name):
    return references[name]

def load_reference(name_or_id):
    if name_or_id in references:
        return load(references[name_or_id])
    else:
        return load(name_or_id)


```

Con este codigo GIT puede usar nombres legigles para humanos como "master" para referirse a un snapshot particular en la historia, en lugar de una cadena larga hexadecimal. Tambien si se desea saber en que momento estamos actualmente en la historia, podemos saberlo y asi como resultado si se desea realizar un nuevo snapshot sepamos a que se refiere.
